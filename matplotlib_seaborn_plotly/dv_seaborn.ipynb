# Data Visualization with Seaborn

- Import Libraries
- Read data
- Scatterplot
- Boxplot
- Violinplot
- Heatmap
- Histogram
- Countplot
- Jointplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='unicode-escape')
df.head()
```

```python
df[df['Legendary'] == True]
```

```python
sns.lmplot(x='Attack', y='Defense', data=df)
plt.show()
```

```python
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
plt.show()
```

```python
df_copy = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
sns.boxplot(data=df_copy)
plt.show()
```

```python
sns.violinplot(data=df_copy)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='Type 1', y='Attack', data=df_copy)
plt.show()
```

```python
corr = df_copy.corr()

sns.heatmap(corr)
plt.show()
```

```python
sns.distplot(df.Attack, color='red')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Type 1', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.jointplot(df.Attack, df.Defense, kind='kde', color='red')
plt.show()
```
